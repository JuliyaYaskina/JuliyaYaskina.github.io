```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
salaries = np.array(df['Зарплата'])
# Расчет параметров
mean = np.mean(salaries)
std = np.std(salaries)

# Создание теоретической кривой
x = np.linspace(mean - 4*std, mean + 4*std, 1000)
pdf = norm.pdf(x, mean, std)

# Построение графика
plt.plot(x, pdf, 'r-', linewidth=2)
plt.title('Нормальное распределение зарплат')
plt.xlabel('Зарплата')
plt.ylabel('Плотность')
plt.show()
```

```python
# Вычисление статистических характеристик
mean_salary = np.mean(salaries)
median_salary = np.median(salaries)
variance = np.var(salaries)
std_dev = np.std(salaries)

# Вывод результатов
print(f"Основные статистические характеристики:")
print(f"Среднее значение: {mean_salary:.2f}")
print(f"Медиана: {median_salary:.2f}")
print(f"Дисперсия: {variance:.2f}")
print(f"Стандартное отклонение: {std_dev:.2f}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Группировка по городам и расчет описательных статистик
stats = df.groupby('Регион')['Зарплата'].agg([
    'mean',        # среднее
    'median',      # медиана
    'std',         # стандартное отклонение
    'var',         # дисперсия
    'min',         # минимум
    'max',         # максимум
    'count'        # количество наблюдений
]).round(2)

print("Описательные статистики по городам:")
print(stats)

# Визуализация различий
plt.figure(figsize=(14, 8))

# Boxplot для сравнения распределений
plt.subplot(1, 2, 1)
sns.boxplot(x='Регион', y='Зарплата', data=df, palette='viridis')
plt.title('Сравнение зарплат по городам')
plt.ylabel('Зарплата')
plt.xlabel('Регион')

# График плотности распределения
plt.subplot(1, 2, 2)
sns.kdeplot(data=df, x='Зарплата', hue='Регион', fill=True, alpha=0.4)
plt.title('Плотность распределения зарплат')
plt.xlabel('Регион')
plt.ylabel('Плотность')

plt.tight_layout()
plt.show()

# Оценка разницы между средними значениями
mean_diff = stats['mean'].max() - stats['mean'].min()
print(f"\nРазница между максимальной и минимальной средней зарплатой: {mean_diff:.2f}")

# Оценка разницы между медианными значениями
median_diff = stats['median'].max() - stats['median'].min()
print(f"Разница между максимальной и минимальной медианной зарплатой: {median_diff:.2f}")
```

```python
def get_mode(df):
    return df.mode()[0] if not df.empty else None

# Группируем данные по городам и находим моду для каждой группы
mode_by_city = df.groupby('Регион')['Профессия'].apply(get_mode)

print("Мода по профессиям в городах:")
print(mode_by_city)
```

```python
mean_salaries = df.groupby('Уровень образования')['Зарплата'].mean().sort_values(ascending=False)

print("Средние зарплаты по уровням образования:")
print(mean_salaries)

# Находим уровень образования с максимальной средней зарплатой
max_salary_education = mean_salaries.idxmax()
max_salary_value = mean_salaries.max()

print(f"\nМаксимальная средняя зарплата у работников с образованием: {max_salary_education}")
print(f"Средняя зарплата: {max_salary_value:.2f} рублей")
```

```python
male_df = df[df['Пол'] == 'мужской']

# Группируем по профессиям и считаем средние зарплаты
avg_salaries = male_df.groupby('Профессия')['Зарплата'].mean().sort_values(ascending=False)

print("Средние зарплаты мужчин по профессиям:")
print(avg_salaries)

# Находим профессию с максимальной средней зарплатой
max_salary_profession = avg_salaries.idxmax()
max_salary_value = avg_salaries.max()

print(f"\nМаксимальная средняя зарплата у мужчин в профессии: {max_salary_profession}")
print(f"Средняя зарплата: {max_salary_value:.2f} рублей")

# Визуализация результатов
plt.figure(figsize=(14, 8))

# График средних зарплат по профессиям
plt.subplot(1, 2, 1)
sns.barplot(
    x=avg_salaries.values,
    y=avg_salaries.index,
    palette='viridis',
    order=avg_salaries.sort_values(ascending=False).index
)
plt.title('Средние зарплаты мужчин по профессиям')
plt.xlabel('Средняя зарплата (руб.)')
plt.ylabel('Профессия')

# График распределения зарплат по профессиям
plt.subplot(1, 2, 2)
sns.boxplot(
    x='Зарплата',
    y='Профессия',
    data=male_df,
    order=avg_salaries.sort_values(ascending=False).index,
    palette='magma'
)
plt.title('Распределение зарплат мужчин по профессиям')
plt.xlabel('Зарплата (руб.)')
plt.ylabel('Профессия')

plt.tight_layout()
plt.show()

# Дополнительный анализ
print("\nДетальный анализ распределения зарплат:")
print(male_df.groupby('Профессия')['Зарплата'].describe())
```

```python
female_df = df[df['Пол'] == 'женский']

# Группируем по регионам и считаем медианные зарплаты
median_salaries = female_df.groupby('Регион')['Зарплата'].median().sort_values()

print("Медианные зарплаты женщин по регионам:")
print(median_salaries)

# Находим регион с минимальной медианной зарплатой
min_median_region = median_salaries.idxmin()
min_median_value = median_salaries.min()

print(f"\nМинимальная медианная зарплата у женщин в регионе: {min_median_region}")
print(f"Медианная зарплата: {min_median_value:.2f} рублей")

# Визуализация результатов
plt.figure(figsize=(14, 8))

# График медианных зарплат по регионам
plt.subplot(1, 2, 1)
sns.barplot(
    x=median_salaries.index,
    y=median_salaries.values,
    palette='coolwarm',
    order=median_salaries.sort_values().index
)
plt.title('Медианные зарплаты женщин по регионам')
plt.xlabel('Регион')
plt.ylabel('Медианная зарплата (руб.)')
plt.xticks(rotation=45)

# График распределения зарплат по регионам
plt.subplot(1, 2, 2)
sns.boxplot(
    x='Регион',
    y='Зарплата',
    data=female_df,
    order=median_salaries.sort_values().index,
    palette='magma'
)
plt.title('Распределение зарплат женщин по регионам')
plt.xlabel('Регион')
plt.ylabel('Зарплата (руб.)')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

# Дополнительный анализ
print("\nДетальный анализ распределения зарплат:")
print(female_df.groupby('Регион')['Зарплата'].describe())
```

```python
ekb_df = df[df['Регион'] == 'Екатеринбург']

# Создаем возрастные группы
age_groups = pd.cut(
    ekb_df['Возраст'],
    bins=[0, 19, 30, 45, 60, np.inf],
    labels=['до 19', '20-30', '31-45', '46-60', 'после 60']
)

# Добавляем колонку с возрастными группами
ekb_df['Возрастная группа'] = age_groups

# Группируем данные
grouped = ekb_df.groupby(['Возрастная группа', 'Уровень образования'])['Зарплата'].median().reset_index()

# Находим максимальную среднюю зарплату
max_salary = grouped['Зарплата'].max()
max_conditions = grouped[grouped['Зарплата'] == max_salary]

print("Максимальная средняя зарплата:")
print(max_conditions)

# Визуализация
plt.figure(figsize=(16, 10))

# Тепловая карта средних зарплат
plt.subplot(1, 2, 1)
sns.heatmap(
    grouped.pivot_table(
        values='Зарплата',
        index='Уровень образования',
        columns='Возрастная группа',
        aggfunc='mean'
    ),
    annot=True,
    fmt='.0f',
    cmap='coolwarm',
    cbar_kws={'label': 'Средняя зарплата (руб)'}
)
plt.title('Средние зарплаты по образованию и возрасту')
plt.xlabel('Возрастная группа')
plt.ylabel('Уровень образования')

# График средних зарплат
plt.subplot(1, 2, 2)
sns.barplot(
    x='Зарплата',
    y='Уровень образования',
    hue='Возрастная группа',
    data=grouped,
    palette='viridis'
)
plt.title('Средние зарплаты по образованию и возрасту')
plt.xlabel('Средняя зарплата (руб)')
plt.ylabel('Уровень образования')
plt.legend(title='Возрастная группа')

plt.tight_layout()
plt.show()

# Вывод результатов
print("\nСредние зарплаты по группам:")
print(grouped.pivot_table(
    values='Зарплата',
    index='Уровень образования',
    columns='Возрастная группа',
    aggfunc='mean'
).round(2))
```
